```python
df=df.drop_duplicates()
```

```python
#drop date column 
df=df.drop(columns=['Date_Collected'])
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
#label encode the columns ,'Difficulty','TrailMarkersInstalled'
# could do these later if needed: 'Width_ft', 'Class', 'Surface', 'Gen_Topog'
label_encoder = LabelEncoder()
columns_to_encode = ['Difficulty', 'TrailMarkersInstalled']

for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])

```

```python
df.head()
```

```python
#break shape into 2 new columns
# first number in shape latitude and second is longitude
df[['Latitude', 'Longitude']] = df['SHAPE'].str.split(' ').str[1:3].apply(lambda x: pd.Series([x[0], x[1]]))
```

```python
df.head()
```

```python
df['Latitude'] = df['Latitude'].str.replace('[,()]', '', regex=True)
df['Longitude'] = df['Longitude'].str.replace('[,()]', '', regex=True)
```

```python
df.drop(columns=['SHAPE'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#if park name null drop and if park id null drop
df = df.dropna(subset=['Park_Name', 'ParkID'])
```

```python
#replace null values with 0 n/a for trail name
df['Trail_Name'] = df['Trail_Name'].fillna('N/A')
```

```python
df.head(10)
```

```python
#if string "Unnamed Official Trail" in trail name replace with N/A
df['Trail_Name'] = df['Trail_Name'].replace('Unnamed Official Trail', 'N/A')
```

```python
#save to csv
df.to_csv('Parks_Trails_cleaned.csv', index=False)
```
